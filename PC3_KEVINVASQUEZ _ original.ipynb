```python
class Nodo:
    def __init__(self,valor):
        self.valor =valor
        self.iz=None
        self.der= None
        
raiz=Nodo("+")
raiz.iz= Nodo("-")
raiz.der= Nodo("*")
raiz.iz.iz=Nodo("*")
n20= raiz.iz.der=Nodo(20)
raiz.iz.iz.iz= Nodo("+")
n3= raiz.iz.iz.der= Nodo(3)
n10= raiz.iz.iz.iz.iz= Nodo(10)
n1= raiz.iz.iz.iz.der= Nodo(1)
n3b= raiz.der.iz= Nodo(3)
n2= raiz.der.der= Nodo(2)



#recorrido inorden

def inorden(raiz):
    if raiz:
        inorden(raiz.iz)
        print(raiz.valor,end=" ")
        inorden(raiz.der)
        
print("\n RECORRIDO IN-ORDEN")
resultado = (n10.valor + n1.valor)* n3.valor - n20.valor + (n3b.valor * n2.valor)
inorden(raiz)
print("\n-----------")
print("resultado es",resultado)
    
 #recorrido preorden       
def preorden(raiz):
    if raiz:
        print(raiz.valor,end=" ")
        inorden(raiz.iz)
        inorden(raiz.der)
print("\n RECORRIDO PRE-ORDEN")


preorden(raiz)       
 #recorrido postorden       
def postorden(raiz):
    if raiz:
        inorden(raiz.iz)
        inorden(raiz.der)
        print(raiz.valor,end=" ")
print("\n RECORRIDO Post-ORDEN")
postorden(raiz)  



def contar_nodos(r):
    if r is None:
        return 0
    return 1 + contar_nodos(r.iz) + contar_nodos(r.der)

print("\n.................")
print("CANTIDAD DE NODOS ES :", contar_nodos(raiz), "nodos")
#evaluar

#(10+1)
print(n10.valor+n1.valor)

# 11* 3
print(contar_nodos(raiz)* n3.valor)

# 33 -20   (numero concatenado en el 33 del nodo izquierdo y derecho)
concatenar= (str(n3.valor)+str(n3b.valor))
num33= int(concatenar)
print(num33-20)

# 3 *2
print (n3b.valor * n2.valor)

# 13 + 6
cont =(str(n1.valor)+str(n3.valor))
num13 = int (cont)
print(num13 + 6)







        


			
```
