```python
class Nodo:
    def __init__(self,valor):
        self.valor =valor
        self.iz=None
        self.der= None
        
raiz=Nodo("+")
raiz.iz= Nodo("-")
raiz.der= Nodo("*")
raiz.iz.iz=Nodo("*")
n20= raiz.iz.der=Nodo(20)
raiz.iz.iz.iz= Nodo("+")
n3= raiz.iz.iz.der= Nodo(3)
n10= raiz.iz.iz.iz.iz= Nodo(10)
n1= raiz.iz.iz.iz.der= Nodo(1)
n3b= raiz.der.iz= Nodo(3)
n2= raiz.der.der= Nodo(2)



#recorrido inorden

def inorden(raiz):
    if raiz:
        inorden(raiz.iz)
        print(raiz.valor,end=" ")
        inorden(raiz.der)
        
print("\n RECORRIDO IN-ORDEN")
resultado = (n10.valor + n1.valor)* n3.valor - n20.valor + (n3b.valor * n2.valor)
inorden(raiz)
print("\n-----------")
print("resultado es",resultado)
    
 #recorrido preorden       
def preorden(raiz):
    if raiz:
        print(raiz.valor,end=" ")
        inorden(raiz.iz)
        inorden(raiz.der)
print("\n RECORRIDO PRE-ORDEN")


preorden(raiz)       
 #recorrido postorden       
def postorden(raiz):
    if raiz:
        inorden(raiz.iz)
        inorden(raiz.der)
        print(raiz.valor,end=" ")
print("\n RECORRIDO Post-ORDEN")
postorden(raiz)    







        


			
```
